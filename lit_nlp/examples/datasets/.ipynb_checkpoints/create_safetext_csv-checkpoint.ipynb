```python
import pandas as pd

df = pd.read_json('./SafeText/SafeText.json', orient='index') # sets scenario as index
df.reset_index(inplace=True)
df = df.rename(columns = {'index':'scenario'})
```

```python
df
```

```python
df.to_csv('./SafeText/safetext_for_generation.csv', index=False)  
```

```python
data = {'scenario': [], 'advice': [], 'label': []}
for _, row in df.iterrows():
    scenario = row['scenario']
    for advice in row['safe']:
        data['scenario'].append(scenario)
        data['advice'].append(advice)
        data['label'].append('safe')
        
    for advice in row['unsafe']:
        data['scenario'].append(scenario)
        data['advice'].append(advice)
        data['label'].append('unsafe')
```

```python
df2 = pd.DataFrame(data)
```

```python
df2
```

```python
df2.to_csv('./SafeText/safetext_for_decision.csv', index=False)  
```
